# 공간 필터링
## 평균 필터링

```python
img = cv2.imread('/content/gdrive/My Drive/디지털 영상처리 강의 관련/Images/lena256.jpg')
#print(img.shape)
#cv2_imshow(img)
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#plt.imshow(RGB_img)

gimg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(gimg.shape)
cv2_imshow(gimg)
plt.imshow(gimg, cmap='gray')
```

```python
#평균 필터 만들기
kernal3 = np.ones((3, 3), np.float32) / 9 # 각 계수가 1/9인 필터
kernal5 = np.ones((5, 5), np.float32) / 25
```

```python
outimg3 = cv2.filter2D(gimg, -1, kernal3) # -1 은 입력과 동일한 형태
cv2_imshow(outimg3)
print("3x3 커널 출력")
```

```python
outimg5 = cv2.filter2D(gimg, -1, kernal5) # -1 은 입력과 동일한 형태
cv2_imshow(outimg5)
print("5x5 커널 출력")
```

```python
outimg33 = cv2.filter2D(outimg3, -1, kernal3) # 3x3 한 바퀴 더 돌린 이미지
cv2_imshow(outimg33)
```

```python
outimg333 = cv2.filter2D(outimg33, -1, kernal3) # 3x3 두 바퀴 더 돌린 이미지
cv2_imshow(outimg333)
# 커널을 여러번 돌리면 blur 효과가 강해짐을 확인
```

직접 커널 만들기

```python
kernal3[0][0] = 1/9
kernal3[0][1] = 1/9
kernal3[0][2] = 1/9
kernal3[1][0] = 1/9
kernal3[1][1] = 1/9
kernal3[1][2] = 1/9
kernal3[2][0] = 1/9
kernal3[2][1] = 1/9
kernal3[2][2] = 1/9
```

```python
outimg3 = cv2.filter2D(gimg, -1, kernal3) # -1 은 입력과 동일한 형태
cv2_imshow(outimg3)
```

원본 이미지를 주는 커널

```python
kernal3[0][0] = 0
kernal3[0][1] = 0
kernal3[0][2] = 0
kernal3[1][0] = 0
kernal3[1][1] = 1
kernal3[1][2] = 0
kernal3[2][0] = 0
kernal3[2][1] = 0
kernal3[2][2] = 0
```

```python
outimg3 = cv2.filter2D(gimg, -1, kernal3) # -1 은 입력과 동일한 형태
cv2_imshow(outimg3)
```

```python
kernal3[0][0] = -1/9
kernal3[0][1] = -1/9
kernal3[0][2] = -1/9
kernal3[1][0] = -1/9
kernal3[1][1] = 8/9
kernal3[1][2] = -1/9
kernal3[2][0] = -1/9
kernal3[2][1] = -1/9
kernal3[2][2] = -1/9

outimg3 = cv2.filter2D(gimg, -1, kernal3) # -1 은 입력과 동일한 형태
cv2_imshow(outimg3)
```
